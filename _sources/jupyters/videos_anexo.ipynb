Vamos a ver ahora cómo montamos el circuito en qiskit

````python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, DensityMatrix
````

````python
# 8 es una potencia de 2
psi = Statevector.from_int(0,8)
psi.draw('latex')
````

````python
qc = QuantumCircuit(3)
qc.h([0,1])
qc.cx(0,1)
qc.cx(0,2)
qc.draw('mpl')
````

````python
psiout = psi.evolve(qc)
psiout.draw('latex')
````

````python
from qiskit.tools.visualization import plot_state_city
````

````python
plot_state_city(psiout)
````

````python
# También se puede sacar así
psiout.draw('city')
````

NOTA: El gráfico anterior (*plot_state_city*) se obtiene de la matriz de densidad que se saca a continuación. Como puede verse, las barras del gráfico tienen como altura los valores de esa matriz de densidad.

````python
rhoout = DensityMatrix(psiout)
rhoout.draw('latex')
````

## Diferencia puertas cuánticas.

Esta sería la pregunta 1.5 del examen de qiskit 2022. Está en la siguiente dirección

<a href="https://www.youtube.com/watch?v=nPRDL5Dp9-g" target="_blank"> https://www.youtube.com/watch?v=nPRDL5Dp9-g </a>

![](images/pregunta1_5_1.PNG)

La matriz $U_1(\lambda)$ se define como:

$$U_{1}(\lambda)=\left[\begin{array}{cc}
1 & 0\\
0 & e^{i\lambda}
\end{array}\right]\longrightarrow P(\lambda)=\left[\begin{array}{cc}
1 & 0\\
0 & e^{i\lambda}
\end{array}\right]$$

Entonces, se tiene lo siguiente

$$U_{1}(\pi)=\left[\begin{array}{cc}
1 & 0\\
0 & e^{i\pi}
\end{array}\right]\longrightarrow P(\pi)=\left[\begin{array}{cc}
1 & 0\\
0 & e^{i\pi}
\end{array}\right]=\left[\begin{array}{cc}
1 & 0\\
0 & ^{-1}
\end{array}\right]$$

Por otra parte:

$$U_{2}(\varPhi,\lambda)=R_{z}(\varPhi)R_{y}(\pi/2)R_{z}(\lambda)$$

$$U_{2}(\varPhi,\lambda)=\frac{1}{\sqrt{2}} \left[\begin{array}{cc}
1 & -e^{i\lambda}\\
e^{i\varPhi} & e^{i(\varPhi+\lambda)}
\end{array}\right]$$

$$U_{2}(0,\pi)=\frac{1}{\sqrt{2}}\left[\begin{array}{cc}
1 & 1\\
1 & -1
\end{array}\right]\longrightarrow Puerta\ H$$

![](images/pregunta1_5_2.PNG)

Con quiskit lo hariamos así

````python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Operator
from qiskit.tools.visualization import array_to_latex
import numpy as np

````

````python
qc=QuantumCircuit(1)
qc.u1(np.pi,0)
qc.draw('mpl')
````

````python
Ulop = Operator(qc)
array_to_latex(Ulop)
````

````python
qc=QuantumCircuit(1)
qc.p(np.pi,0)
qc.draw('mpl')
````

````python
Pop = Operator(qc)
array_to_latex(Pop)
````

````python
qc=QuantumCircuit(1)
qc.u2(0,np.pi,0)
qc.draw('mpl')
````

````python
U2op = Operator(qc)
array_to_latex(U2op)
````

````python
qc=QuantumCircuit(1)
qc.u(np.pi/2,0,np.pi,0)
qc.draw('mpl')
````
